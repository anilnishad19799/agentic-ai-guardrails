## Two method using LLM

### First using Call directly in guard

```python
from guardrails import Guard
import os
```

```python
guard = Guard()
```

```python
res = guard(
    model="gpt-3.5-turbo",
    messages=[{
        "role": "user",
        "content": "How do I make a cake?"
    }]
)
```

```python
print(res.raw_llm_output)
print(res.validated_output)
print(res.validation_passed)
```

### Parse after llm output

```python
import openai
from guardrails import Guard
```

```python
guard = Guard()
```

```python
output = openai.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{
        "role": "user",
        "content": "How do I make a cake?"
    }]
).choices[0].message.content
```

```python
res = guard.parse(
    llm_output=output
)
```

```python
print(res.validated_output) # Validated output
```

### With reask

```python
output = openai.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{
        "role": "user",
        "content": "How do I make a cake?"
    }]
).choices[0].message.content
```

```python
res = guard.parse(
    llm_output=output,
    model="gpt-3.5-turbo",
    num_reasks=1
)

print(res.validated_output) # Validated output
print(guard.history.last.reasks) # A list of reasks
```
