```python
from guardrails import Guard
# from guardrails.hub import DetectPII
from guardrails.hub import DetectPII
```

```python
guard = Guard().use(
    DetectPII(pii_entities=["EMAIL_ADDRESS", "PHONE_NUMBER", "CREDIT_CARD", "SOCIAL_SECURITY_NUMBER"], on_fail="fix")
    )
```

```python
input_str = "contact me at anilnishad@gmail.com"
validated = guard.validate(input_str)
print(validated)
```

```python
print(validated.validated_output)
```
